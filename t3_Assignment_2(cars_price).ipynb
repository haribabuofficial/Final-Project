# split the data into feature and target

```python
X = df10[['horsepower', 'curb-weight', 'engine-size', 'highway-kmpl', 'wheel-base', 'city-kmpl', 'length', 'width', ]]

y = df10['price']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

# Tune the parameters of models for Random Forest Regressor

```python
from sklearn.tree import DecisionTreeRegressor

from sklearn.model_selection import KFold, GridSearchCV

kf = KFold(n_splits = 5)

model = DecisionTreeRegressor()

params = {"criterion": ["squared_error", "friedman_mse", "absolute_error"],
          "splitter": ["best", "random"],
          "max_features": [2, 4, 6, 8, None],
          "max_depth": [3, 4, 5, 6]}

gcv = GridSearchCV(model, param_grid = params, cv = kf, scoring = "r2")

gcv.fit(x_train, y_train)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
gcv.best_estimator_
```

```python
pred = gcv.best_estimator_.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, pred)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, pred)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, pred)
```
